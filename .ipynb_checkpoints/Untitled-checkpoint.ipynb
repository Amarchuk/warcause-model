```python
dfw['yng'] = np.nan
dfw['old'] = np.nan

ind1 = []
for area in common_areas:
    print area
    tmpdw = dfw[dfw.COUNTRY == area]
    tmpda = dfa[dfa.area == area]
    ma, mi = tmpda.refdate.max(), tmpda.refdate.min()
    for index, row in tmpdw.iterrows():
        if row.YEAR < mi or row.YEAR > ma:
            ind1.append((area, row.YEAR))
        else:
            year = row.YEAR
            ra = np.linspace(tmpda[tmpda.refdate == (year - year%5)][['10-14']].iat[0,0], 
                             tmpda[tmpda.refdate == (year - year%5 + 5)][['10-14']].iat[0,0], 5)
            dfw.set_value(index, 'yng', ra[year%5])
            ra = np.linspace(tmpda[tmpda.refdate == (year - year%5)][['20-24']].iat[0,0], 
                             tmpda[tmpda.refdate == (year - year%5 + 5)][['20-24']].iat[0,0], 5)
            dfw.set_value(index, 'old', ra[year%5])

for a,y in ind1:
    dfw = dfw.drop(dfw[dfw.COUNTRY == a][dfw.YEAR == y].index)

          
print dfw[0:3]
```

```python
len(dfw)
```

```python
dfw['summ'] = dfw['yng'] + dfw['mid'] + dfw['old']
```

```python
fig, axes = plt.subplots(1, ncols=4, figsize=(16, 5))
axes[0].plot(dfw.ACTOTAL, dfw.yng, '.')
axes[1].plot(dfw.ACTOTAL, dfw.mid, '.')
axes[2].plot(dfw.ACTOTAL, dfw.old, '.')
axes[3].plot(dfw.ACTOTAL, dfw.summ, '.')
plt.show()
```

```python
corr_group_by_area = []
for area in common_areas:
    tmpdw = dfw[dfw.COUNTRY == area]
    corr = np.corrcoef(tmpdw.ACTOTAL, tmpdw.mid)[1, 0]
    print '%s : %s' % (area, corr)
    corr_group_by_area.append((corr, area))
```

```python
corr_group_by_area = [x for x in corr_group_by_area if pd.notnull(x[0])]
```
